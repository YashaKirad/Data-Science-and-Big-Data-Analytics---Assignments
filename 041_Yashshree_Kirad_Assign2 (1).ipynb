```python
min_value = df['GradeID'].min()
max_value = df['GradeID'].max()
print(min_value)
print(max_value)
```

```python
bins = np.linspace(min_value,max_value,4)
bins
```

```python
labels = ['Okay', 'Good', 'Great']
```
