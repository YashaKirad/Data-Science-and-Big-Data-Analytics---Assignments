```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("virus.csv" )
```

```python
df
```

```python
min_value_V = df['V1'].min()
max_value_V = df['V1'].max()
print(min_value_V)
print(max_value_V)
```

```python
#equal width binning
```

```python
labe=["A","B","C","D","E"]
```

```python
df['width']=pd.cut(df["V1"],5,labels=labe)
```

```python
df
```

```python
#equal freq
la=["1","2","3","4","5"]
df['freq']=pd.qcut(df["V1"],q=5,labels=la)
```

```python
df
```

```python
cat=["low","high"]
bins = np.linspace(min_value_V,max_value_V,3)
df['group'] = pd.cut(df['V1'], bins=bins, labels=cat, include_lowest=True)
df
```

```python
grouped=df.groupby(df.group)
g1=grouped.get_group("low")
g2=grouped.get_group("high")
```

```python
print(g1)
```

```python
print(g2)
```

```python
mean_g1=g1["V1"].mean()
print(mean_g1)
```

```python
mean_g2=g2["V1"].mean()
print(mean_g2)
```

```python
median_g1=g1["V1"].median()
print(median_g1)
```

```python
median_g2=g2["V1"].median()
print(median_g2)
```

```python
std_g1=g1["V1"].std()
print(std_g1)
```

```python
std_g2=g2["V1"].std()
print(std_g2)
```

```python
q1_g1=g1["V1"].quantile(q=0.25)
print(q1_g1)
```

```python
q1_g2=g2["V1"].quantile(q=0.25)
print(q1_g2)
```

```python
q3_g1=g1["V1"].quantile(q=0.75)
print(q3_g1)
```

```python
q3_g2=g2["V1"].quantile(q=0.75)
print(q3_g2)
```

```python
plt.boxplot(g1['V1'])
plt.title('Box plot of v1 in group 1')
plt.ylabel('V1')
plt.show()
```

```python
plt.boxplot(g2['V1'])
plt.title('Box plot of v1 in group 2')
plt.ylabel('V1')
plt.show()
```

```python
g3=g1.copy()
g4=g2.copy()
print(g3)
print('\n')
print(g4)
```

```python
g3['ANTI_A']= (g3['ANTI_A']-g3['ANTI_A'].min())/(g3['ANTI_A'].max()-g3['ANTI_A'].min())
g3['ANTI_B']= (g3['ANTI_B']-g3['ANTI_B'].min())/(g3['ANTI_B'].max()-g3['ANTI_B'].min())

print(g3)
```

```python
g4['ANTI_A']= (g4['ANTI_A']-g4['ANTI_A'].min())/(g4['ANTI_A'].max()-g4['ANTI_A'].min())

g4['ANTI_B']= (g4['ANTI_B']-g4['ANTI_B'].min())/(g4['ANTI_B'].max()-g4['ANTI_B'].min())

print(g4)
```
